## 1. Modificando la Leyenda.

Comenzaremos con el último gráfico de la parte 1. Los colores de los puntos y su tamaño están determinados por las variables 'state' (categórica) y 'popdensity' (continua), respectivamente.

Aquí tenemos lo esencial: título, ejes etiquetados y leyenda.
Con la función *theme()* podremos refinar más detalles visuales en nuestro gráfico.

```python
# cargando modulo necesario
from plotnine import *
import matplotlib.pyplot as plt
import pandas as pd   
midwest = pd.read_csv("http://goo.gl/G1K41K") # descargando el dataset
```

```python
gg = ggplot(midwest, aes(x='area', y='poptotal', color = 'state')) +\
xlim(0,0.1) + ylim(0,1000000)+\
geom_point(aes(size='popdensity'), na_rm= True) +\
scale_color_brewer(type = 'qual',palette = 'Set1') +\
labs(title="Área Vs Población", y="Población", x="Área")
gg
```

En plotnine podemos modificar elementos del gráfico con los siguientes comandos:

1. element_line
2. element_rect
3. element_text

Cada uno para modificar diferentes partes del gráfico, siendo estas: lineas, rectángulos (por ejemplo en la leyenda), y todo lo referente a texto (títulos, etiquetas en los ejes, etc.)

Ejemplos de este último a continuación:

```python
gg + theme(title=element_text(size=20, 
                                    face="bold", 
                                    family="American Typewriter",
                                    color="tomato",
                                    hjust=0.5,
                                    lineheight=1.2),
           legend_title=element_text(size = 15, color='blue'),# title
            axis_title_x=element_text(vjust=10,  
                                      size=15),  # X axis title
            axis_title_y=element_text(size=15),  # Y axis title
            axis_text_x=element_text(size=10, 
                                     angle = 30,
                                     vjust=.5),  # X axis text
            axis_text_y=element_text(size=10))  # Y axis text
```

Podemos además cambiar las categorías que aparecen en la leyenda, entregando de mejor manera la información del gráfico, e incluso ocupar colores específicos para cada categoría presente.

```python
gg + scale_color_manual(name="State", 
                        labels = ["Illinois", 
                                   "Indiana", 
                                   "Michigan", 
                                   "Ohio", 
                                   "Wisconsin"], 
                        values = ["blue", 
                                   "red", 
                                   "green", 
                                   "brown", 
                                   "orange"])
```
